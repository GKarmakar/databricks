```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('mylogreg').getOrCreate()
```

```
from pyspark.ml.classification import LogisticRegression
```

```
my_data = spark.read.format('libsvm').load('/FileStore/tables/sample_libsvm_data.txt')
```

```
my_data.show()
```

```
my_logreg_model = LogisticRegression()
```

```
fitted_logreg_model = my_logreg_model.fit(my_data)
```

```
logreg_summary = fitted_logreg_model.summary
```

```
logreg_summary.predictions.printSchema()
```

```
logreg_summary.predictions.show()
```

```
lr_train, lr_test = my_data.randomSplit([0.7, 0.3])
```

```
final_model = LogisticRegression()
```

```
fitted_lr = final_model.fit(lr_train)
```

```
prediction_and_labels = fitted_lr.evaluate(lr_test)
```

```
prediction_and_labels.predictions.show()
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
```

```
prediction_and_labels.predictions.show()
```

```
my_eval = BinaryClassificationEvaluator()
```

```
my_final_roc = my_eval.evaluate(prediction_and_labels.predictions)
```
